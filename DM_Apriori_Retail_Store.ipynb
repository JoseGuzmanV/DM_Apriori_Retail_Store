```python
# Importa librerias importantes
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

data = pd.read_csv('/content/drive/My Drive/LIBROS/MINERIA DE DATOS/DM_Sales_Extract.csv', header=0, index_col=False)
data
```

```python
# Convertir columna data['Products'] en LISTA
data['Products'] = data['Products'].apply(lambda x: x.split(','))

# data['Products']` column into a list of strings
#data['Products'] = data['Products'].apply(lambda x: ' '.join(map(str, x)))
```

```python
association_rules = apriori(data['Products'], min_support=0.0045, min_confidence=0.2, min_lift=2.5, max_length=4)
association_results = list(association_rules)
association_results
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
# Por cada regla
for item in association_results:

    # Primer elemento de la regla contiene el par [items base, item añadido]
    pair = item[0]
    items = [x for x in pair]
    print("Regla: {}  ->  {}".format( items[:-1],  items[-1] ) ) # imprime items base y item añadido (el ultimo de items)

    # Segundo elemento de la regla es el soporte
    print("Soporte: " + str(item[1]))

    # Tercer elemento de la regla es son estadisticas (Confianza, Lift)
    print("Confianza: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python

#**Lift:**
#- Es una medida de la correlación entre dos eventos.
#- Se calcula dividiendo el soporte de la regla por el producto de los soportes de los elementos individuales de la regla.
#- Un valor de lift mayor que 1 indica que los dos eventos son más propensos a ocurrir juntos que si fueran independientes.
#- Un valor de lift menor que 1 indica que los dos eventos son menos propensos a ocurrir juntos que si fueran independientes.

#**Soporte:**
#- Es una medida de la frecuencia con la que ocurre una regla.
#- Se calcula dividiendo el número de transacciones que contienen la regla por el número total de transacciones.
#- Un valor de soporte alto indica que la regla es común.
#- Un valor de soporte bajo indica que la regla es poco común.

#**Confianza:**
#- Es una medida de la probabilidad de que ocurra el elemento añadido dado que los elementos base también ocurren.
#- Se calcula dividiendo el soporte de la regla por el soporte de los elementos base.
#- Un valor de confianza alto indica que el elemento añadido es muy probable que ocurra dado que los elementos base también ocurren.
#- Un valor de confianza bajo indica que el elemento añadido no es muy probable que ocurra dado que los elementos base también ocurren.
```
